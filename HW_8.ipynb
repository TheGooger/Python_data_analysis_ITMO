#Работа с API

```python
import requests
import pandas as pd
```

##Информация обо всех книгах

```python
api_base_url = 'https://www.anapioficeandfire.com/api'
```

Так как я не знаю, сколько книг в серии, проверим это, используя пагинацию страниц.

```python
counter = 0
while True:
  try:
    book_i = requests.get(
        f'{api_base_url}/books/{counter+1}'
    ).json()
    counter += 1
  except:
    break
print(f'Всего книг: {counter}')
```

```python
book_list = requests.get(
    f'{api_base_url}/books?pageSize={counter}'
).json()
```

Проверка:

```python
len(book_list)
```

```python
df_books = pd.DataFrame(book_list)
df_books.head()
```

##Информация обо всех домах

Аналогично с домами.

```python
h_counter = 0
while True:
  try:
    requests.get(
        f'{api_base_url}/houses/{h_counter+1}'
    ).json()
    h_counter += 1
  except:
    break
print(f'Всего домов: {h_counter}')
```

```python
house_list = []
for page in range(1, 10):
  house_list += requests.get(
      f'{api_base_url}/houses?page={page}&pageSize=50'
  ).json()
```

```python
len(house_list)
```

```python
df_houses = pd.DataFrame(house_list)
df_houses.head(10)
```
