<br> Столбец <b>price</b> изначально состоит из данных int64, максимальное значение в нем составляет ~10**7, при этом все положительные. Для него предпочтительным типом будет <b>uint32</b>.
<br> Столбец <b>area</b> можно перевести в <b>uint16</b> так как максимальное значение в нем ~20к.
<br> Три следующих int столбца: <b>bedrooms, bathrooms, stories и parking</b> содержат в себе целочисленные данные со значениями до 10, так что меняем на <b>uint8</b>.

Столбцы 6-10 и 12, 13 содержат в себе объекты, среди которых мало уникальных - заменим на <b>category</b>

##Замена типов данных

```python
df_initial = df.copy(deep=True)
```

```python
df["price"] = df["price"].astype(np.uint32)
df["area"] = df["area"].astype(np.uint16)
df["bedrooms"] = df["bedrooms"].astype(np.uint8)
df["bathrooms"] = df["bathrooms"].astype(np.uint8)
df["stories"] = df["stories"].astype(np.uint8)
df["parking"] = df["parking"].astype(np.uint8)
df["mainroad"] = df["mainroad"].astype("category")
df["guestroom"] = df["guestroom"].astype("category")
df["basement"] = df["basement"].astype("category")
df["hotwaterheating"] = df["hotwaterheating"].astype("category")
df["airconditioning"] = df["airconditioning"].astype("category")
df["prefarea"] = df["prefarea"].astype("category")
df["furnishingstatus"] = df["furnishingstatus"].astype("category")
```

```python
df_initial.info(memory_usage="deep")
```

```python
df.info(memory_usage="deep")
```

Потребление памяти было уменьшено в 25 раз.
